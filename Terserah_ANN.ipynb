##### 1. Import library & dataset

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing 
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

df = pd.read_csv('./cancer_classification.csv')
```

#### 2. Data Analysis

```python
# view dimensions of dataset

df.shape
```

```python
# let's preview the dataset

df.head()
```

```python
# view summary of dataset

df.info()
```

```python
# check for missing values in variables

df.isnull().sum()
```

#### 3. Declare feature vector and target variable

```python
X = df.drop(['benign_0__mal_1'], axis=1)

y = df['benign_0__mal_1']
```

#### 4. Split data into test and training set

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

#### 5. Feature Scalling

```python
cols = X_train.columns

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])

X_test = pd.DataFrame(X_test, columns=[cols])
```

#### 6. Building the Artificial Neural Network

```python
# initializing the ANN

model = Sequential()
```

```python
#Adding the input layer and the 1st hidden layer

model.add(Dense(128, activation='relu', input_shape=(30,)))

#Adding the 2nd hidden layer

model.add(Dense(64, activation='relu'))

#Adding the output layer 

model.add(Dense(1, activation='sigmoid'))
```

```python
#Compiling the ANN 

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

#### 7. Training the model

```python
model.fit(X_train, y_train, batch_size=100, epochs=200)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
print('Test accuracy:', round(test_acc,4))
```

#### 8. Data Test

```python
y_pred = model.predict(X_test)
y_pred = y_pred >= 0.1
```

```python
# import metrics to compute accuracy
from sklearn.metrics import accuracy_score

print('Model accuracy score with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
